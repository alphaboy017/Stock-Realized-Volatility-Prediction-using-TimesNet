```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('SP100_return_ffill.csv', index_col=0, parse_dates=True)

# Select 2-3 stocks to visualize (e.g., first three columns)
stocks_to_plot = df.columns[:1]
df[stocks_to_plot].plot(figsize=(12, 6))
plt.title('Returns of Selected SP100 Stocks')
plt.xlabel('Date')
plt.ylabel('Return')
plt.legend(stocks_to_plot)
plt.show()
```
