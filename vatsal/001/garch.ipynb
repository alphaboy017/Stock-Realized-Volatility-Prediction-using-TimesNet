```python
from arch import arch_model
```

```python
data_frame_1=pd.read_csv("SP100_vol_ffill.csv",index_col='Date',parse_dates=True)
```

```python
data_frame_1.iloc[int(len(data_frame_1)*0.7)]
```

```python
returns_SP100.index = pd.to_datetime(returns_SP100.index, format="%d-%m-%Y %H:%M")
```

```python
# Initialize forecast matrix
test_size = int(len(data_frame_1) * 0.3)
num_assets = len(df.columns)
forecasts1 = np.zeros((test_size, num_assets))

# Loop through each asset/column
for i, col in enumerate(df.columns):
    series = returns_SP100[col]

    # Initial model fit to split data correctly
    base_model = arch_model(series, vol='Garch', p=1, o=0, q=1, dist='Normal')
    base_model.fit(update_freq=5, last_obs=pd.Timestamp("2021-12-31"))

    rolling_preds = []

    # Rolling forecast
    for j in range(test_size):
        # Define training window
        end_idx = -(test_size - j)
        train_series = series[:end_idx]

        # Fit GARCH(1,1) model
        model = arch_model(train_series, p=1, o=0, q=1, dist='Normal')
        model_fit = model.fit(disp='off')

        # Forecast 1-step ahead variance
        forecast = model_fit.forecast(horizon=1)
        variance_pred = forecast.variance.values[-1, 0]

        rolling_preds.append(variance_pred)

    # Convert list to array, reshape, and store in forecasts1
    rolling_preds = np.array(rolling_preds).reshape(test_size)
    forecasts1[:, i] = rolling_preds / 100  # scaling to match original logic
```

```python
data_frame_2=data_frame_1.iloc[int(len(data_frame_1)*0.7)+1:len(data_frame_1)]
```

```python
data_frame_2 = data_frame_2.copy()  # Avoid SettingWithCopyWarning

for i, col in enumerate(data_frame_2.columns):
    data_frame_2[col] = forecasts1[:, i]
```

```python
data_frame_2.to_csv("garch_forecasts.csv")
```
